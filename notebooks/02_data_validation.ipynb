# Memory Usage

## Objective

Measure the dataset's memory footprint to understand computational
requirements for analysis and model development.

```python
memory_mb = (
    df.memory_usage(deep=True)
      .sum()
      / 1024**2
)

print(f"Memory Usage: {memory_mb:.2f} MB")
```

### Interpretation

The dataset occupies approximately 120 MB of memory, which is
manageable for exploratory analysis and model development on a modern
personal computer. No immediate optimization or memory reduction
strategies are required.

# Temporal Feature Identification

## Objective

Identify columns that represent timestamps or temporal information for
future time-series feature engineering.

```python
date_columns = [
    col for col in df.columns
    if (
        "date" in col.lower()
        or "time" in col.lower()
        or "epoch" in col.lower()
        or "sunrise" in col.lower()
        or "sunset" in col.lower()
        or "moonrise" in col.lower()
        or "moonset" in col.lower()
    )
]

date_columns
```

### Interpretation

Several variables contain temporal information that may be transformed
into additional predictive features, such as hour of day, day of week,
or seasonal indicators during feature engineering.
